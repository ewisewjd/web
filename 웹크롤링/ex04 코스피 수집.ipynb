## 종목명, 현재가격, 거래량
### 데이터프레임과 csv파일로 제작하기

- 실습 예제
- 주식 (https://finance.naver.com/sise/sise_quant.naver)

```python
# 먼저 필요한 라이브러리(requests)를 임포트 한다
import requests as req
```

```python
# 링크에서 응답을 요청하여 확인
res=req.get("https://finance.naver.com/sise/sise_quant.naver")
```

```python
res.text
```

```python
#컴퓨터 언어로 변환 BeautifulSoup 라이브러리

from bs4 import BeautifulSoup as bs
```

```python
#컴퓨터 언어로 변환
soup=bs(res.text,'lxml')
```

```python
#선택자를 이용하여서 해당 데이터로부터 종목명, 현재가격, 거래량 데이터 가져오기
title=
cuPrice=
chargevalue=
```

```python
#먼저 종목명부터 한번 해볼까?
title=soup.select('a.tltle')
title1=[]
for i in title:
    title1.append(i.text)
title1
```

```python
#다음 현재 가격 만들기
cuPrice=soup.select("td.number:nth-child(3)")
cuPrice1=[]
for i in cuPrice:
    cuPrice1.append(i.text)
cuPrice1
```

```python
# 거래량 데이터 가져오기

chargevalue=soup.select("td.number:nth-child(6)")
chargevalue1=[]
for i in chargevalue:
    chargevalue1.append(i.text)
chargevalue1
```

```python
# 데이터 프레임 만들기 

import pandas as pd
```

```python
title=pd.Series(title1)
title
```

```python
cuPrice=pd.Series(cuPrice1)
cuPrice
```

```python
chargevalue=pd.Series(chargevalue1)
chargevalue
```

```python
stock=pd.concat([title,cuPrice,chargevalue], axis=1)
stock.columns=['종목명','현재가격','거래량']
stock
```

```python
stock.to_csv("주식.csv", encoding='euc-kr')
```

```python
res = req.get('https://finance.naver.com/sise/sise_quant.naver')
soup = bs(res.text, 'lxml')
kospi = pd.DataFrame({
    '종목명' : [],
    '현재가' : [],
    '거래량' : []
})
for n, j in [(2, '종목명'), (3,'현재가'), (6,'거래량')] :
    data = soup.select(f'tr > td:nth-child({n})')
    li = []
    for i in data :
        li.append(i.text)
    kospi[j] = li[5:]
kospi.to_csv('data/kospi.csv', encoding = 'euc-kr')
```
