# 예외처리
- 오류가 나기 전까지 실행하다가 오류가 났을때는 다른 코드를 실행!
- try= 오류나기 전까지 코드
- except= 오류가 났을때 코드



```python
try:
    #오류나기 전
    while True:
        btn=driver.find_element(By.CSS_SELECTOR,"#btn_more > span > a")
        btn.click()
        time.sleep(1)
except:
    
    #오류 난 후
    print("더보기 완료!")
```

# 선택자 한번에 복사해오는 방법
- F12개발자 도구에서 원하는 요소 찾기
- 마우스 우클릭 > copy > copy selector
- [주의점]
- copy selector를 사용시 무조건 선택한 하나만 가지고옴
- 복수개를 요청할때는 선택자의 수정이 필요함
- ex) li:nth-child(3) > a 는 하나만 요청 / 복수개로 변경시에는 li > a 로 수정!

```python
#인수씨 코드 
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
import time
import pandas as pd
#1. 크롬창 실행 후 > 한솥메뉴사이트 요청
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
#2. 더보기를 끝까지 누르는 로직 작성
try :
    while True:
        btn = driver.find_element(By.CSS_SELECTOR,'#btn_more > span > a')
        btn.click()
        time.sleep(1)
except :
    print('더보기 완료')
#3. 가격, 메뉴명을 수집
title = driver.find_elements(By.CLASS_NAME ,'h.fz_03')
price = driver.find_elements(By.CLASS_NAME , 'item-price')
#4. 순수한 텍스트 정보만 추출 -> 비어있는 리스트에 텍스트만 담기
list0 = [(title[i],price[i]) for i in range(len(title))]
list_df = []
for i,j in list0:
    list_df.append((i.text, j.text))
#5. 데이터 프레임 제작 -> 판다스
data = pd.DataFrame(list_df, columns = ['메뉴', '가격'])
print(data)
#6. 한글이 깨지지 않고 csv파일을 제작
data.to_csv('한솥수정본.csv', encoding='euc-kr')
print('완료')
```

# ActionChains
- ActionChains 사용법
    - 1. 내가 액션을 줄 모든 요소들을 수집
    - 2. ActionChains("브라우저")호출
    - 3.  ActionChains("브라우저").기능(요소명).기능(요소명).perform()
        - 여기서 요소명은 내가 만든 변수이름
        - perform()은 전체 실행 

```python
# 이벤트를 묶어서 실행하는 라이브러리 
from selenium.webdriver import ActionChains
```

```python
driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/")
```

```python
# 1.메뉴탭에 마우스 올리기
# 2.올린상태에서 전체 메뉴를 클릭

menu=driver.find_element(By.CSS_SELECTOR,"#gnb > div.gnb_menu > ul > li:nth-child(3) > p > a")

#전체 메뉴를 클릭
allList = driver.find_element(By.CSS_SELECTOR,"#gnb > div.gnb_menu > ul > li:nth-child(3) > div > ul > li:nth-child(1) > a")

#액션체인 쓰기
ActionChains(driver).move_to_element(menu).click(allList).perfrom()
```
